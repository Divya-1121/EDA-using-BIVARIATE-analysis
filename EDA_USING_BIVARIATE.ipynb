# **Heat Map (Categorical - Categorical)**

```python
sns.heatmap(pd.crosstab(iris['size'],iris['fragrance']))
```

```python
(
    iris.groupby('species')['fragrance']
    .value_counts(normalize=True)
    .mul(100)
    .reset_index(name='percentage')
)
```

# **Cluster Map (Categorical - Categorical)**

```python
sns.clustermap(pd.crosstab(iris['species'],iris['fragrance']))
```
